```python
#import pandas library
import pandas as pd
import os
import glob

#list of csv files
all_files = glob.glob(os.path.join(r"C:\Users\pages", "*.csv"))
#create empty dataframe
df_final = pd.DataFrame()
df_list=[]
#loop through list and append to dataframe
for filename in all_files:
    df = pd.read_csv(filename)
    df_list.append(df)
        
df_final = pd.concat(df_list,ignore_index=True)  

df_final.to_csv('all_pages.csv')
```
